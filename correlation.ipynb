## Correlation

* A unitless number between -1 and 1
* Measuring the strength of relationship between two or more variables - their tendency to move in the same direction
* Correlation coefficits closer to extremes indicate a stronger relationship, close to 0 indicate weak or no relationship.
* Generally scatter plots are used to visually inspect for relationships.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yfi
%matplotlib inline
```

```python
stocks = 'META AMZN NFLX GOOG GLD'.split()
data = yfi.download(stocks, start="2020-01-01")['Close']
data = data.resample('BME').last()[:60]
data.plot()

change = pd.DataFrame()
for month in data:
    if month not in change:
        change[month] = np.log(data[month]).diff()
change.head()
corr = change.corr()
corr['AMZN']
```
