## Portfolio Theory

* People have different expectations with respect to risk
* Many people will forego the chance at exceptional returns in favor of "safe"
* The idea behind portfolio theory is that some risk can be mitigated by mixing a number of securities together
* In effect it is possible to earn a higher rate of return while reducing risk

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yfi
%matplotlib inline
```

```python
stocks = 'AAPL GLD'.split()
data = yfi.download(stocks, start="2020-01-01")['Close']

vol = pd.DataFrame()
for stock in stocks:
    if stock not in vol:
        vol[stock] = np.log(data[stock]).diff()
vol = vol[1:]
aapl_var = vol['AAPL'].var() * len(vol)
gld_var = vol['GLD'].var() * len(vol)
print(aapl_var)
print(gld_var)

## Weights and expected ROI from bouth stocks
w_appl = .9
w_gld = 1 - w_appl
exp_aapl = .14
exp_gld = .07
expected_earn = w_appl * exp_aapl + w_gld * exp_gld
print(round(expected_earn,2)*100,"%")
```
