## Valuing Securities

* Valuing stock is an art and science
* Generally you need to expected rate of return and growth
* Basic pricing models - focus on dividend discount model
* There are many models out there, but typically all involve discounting future expectations back to present day
* Once you have the basics, you can build as complex a model as you like

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yfi
%matplotlib inline
```

```python
##Expected return AAPL
div = 2.92
p0 = 152.60
p1 = 178.71

er = round((div + p1) / p0, 2)
print("Expected Return: ", er-1,"%")
pv = (div + p1) / er
print("Present stock value: {}".format(pv))

div = 3.
gr = .08
p1 = 81
er = .12

horizon = pd.DataFrame({'div': np.zeros(3), 'exp_price':np.zeros(3)})
horizon.index = np.arange(1,horizon['div'].count() + 1)
horizon.index.name = 'Year'
for year in range(1, horizon['div'].count() + 1):
    if(year == 1):
        horizon.loc[year, 'div'] = div
        horizon.loc[year, 'exp_price'] = p1
    else:
        horizon.loc[year, 'div'] = div * (1 + gr) ** (year - 1)
        horizon.loc[year, 'exp_price'] = p1 * (1 + gr) ** (year - 1)
horizon

## Stocks with dividents expected return for year 1
year_one = (horizon['div'][1] + horizon['exp_price'][1]) / (er + 1)
year_one

## Stocks without dividends
eps = 5.60 
pe = 15
g = .12
price_to_earnings = eps * pe / (1 + er)
print("Prince_to_earning_value ratio: {}%".format(price_to_earnings))
```
