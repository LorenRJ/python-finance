## Present Value

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
## Basic example

FV = 50000
r = 0.05
T = 1
## Money value form, basic finance, present value = future value / (1 + rate) ** time(years)
PV = FV/(1+r)**T
print(round(PV,2))

## If someone promise us to give 50k the value next year with this rate will be of 47648$

def pv(t, r, fv):
    return round(fv/(1+r) ** t,2)

T = 5
print(pv(T, r, FV))
```

```python
## Example if we won the lotery, 1M spreed out in 20 years

## Gen an array of 20 lenght and 5000 value each cell
cashflows = np.ones(20) * 50000
print(cashflows)
for T in range(len(cashflows)):
    cashflows[T] = cashflows[T]/(1 + r)** T
print(cashflows)    

##Finally make the plot

plt.plot(cashflows)
plt.grid()
plt.xticks([0,5,10,15,20])

## All money earn in this 20 years
print("Sum: ",round(np.sum(cashflows), 2)," $")
```
