### Python For Finance

```python
import numpy as np
import pandas as pd
import numpy.random as npr
```

```python
arr = np.random.standard_normal((10,10))

## Creating data frame from a numpy array
df = pd.DataFrame(arr)

## Call the first 5 rows by default, if you need to show a specific number just set into de function.
df.head(2)

## Call the las 5 lines by default, same logic as the .head method
df.tail()

## Generate fromm a string an array of columns
cols = "A B C D E F G H I J".split()

## Rename the columns names of the original data frame
df.columns = cols

## Rename a single or multiple specific column names
df.rename(columns={'A': 'HOLA', 'C':'Adios'}, inplace= True)

## Return the dataset detail information.
df.info()

## Show the values for a specific column 
df['HOLA']

## Set an increment date time, from the 1/1/2018 to 5/1/2018
index = pd.date_range('1/1/2018', periods=10)

## Set index values into the data frame values, need to provide the index values as the data frame rows number is set
df.set_index(index, inplace= True)

## Accesing data from a row, by their index
df.loc['2018-01-01']

## Accesing data from a specific location, no by index, just the position in the data frame
df.iloc[[1,4]]
```

## Scipy Stats

```python
from scipy import stats as sps
```

```python
## Cumulative area under the normal curve
sps.norm.cdf(-1)

## Generate a list of critical values for various levels of significance
sps.t.isf([.1,.05,.01], [[30],[40]])

## Descriptives
x = sps.norm.rvs(size = 100)
desc = sps.describe(x)
print(desc)

##Percentails
percentails = [5,10,25,75,90,95]
sps.scoreatpercentile(x, percentails)
```

## Matplotlib basics

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(np.random.normal(0,.03, size= 60))

df[1] = df.cumsum()
df.head()
df.plot()

## histogram
x = npr.standard_normal(10000)
plt.hist(x, bins = 50, edgecolor = 'red')

## Define the correlations between various variables with scatter, make a line among the variables independient and dependient.

df6 = pd.DataFrame({'A': npr.normal(size=50), 'B': npr.normal(size=50)})
plt.scatter(df6['A'], df6['B'])
```

## Matplotlib II

```python
import matplotlib.style as mps
```

```python
df = pd.DataFrame({'A': np.random.normal(size=60), 'B': npr.normal(size=60)})
df['C'] = df['B'].cumsum()
x = npr.normal(100, 15, size= 5000)

plt.plot(df.A, df.B)
plt.plot(df.C)
plt.grid(linestyle = ':')
plt.legend(loc=2)
plt.title("This is a nice Plot")
plt.text(20,-1, "This is high")
plt.arrow(26,.17,6,.19)
mps.use('seaborn-v0_8-pastel')

## List predefined styles
##plt.style.available

## typical convention for more a figure made of multiple plots, fig is the entire graphic, ax is a plot within the figure
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.hist(x, bins = 50)
ax2.plot(df.B)

## setting transparency
plt.scatter(df.A, df.B, alpha=.3)
```

```python
from scipy.stats import norm
```

```python
plt.hist(x, bins=50, edgecolor = 'black', density=True)
mean, sd = norm.fit(x)
min_x = float(np.min(x))
max_x = float(np.max(x))
overlay = np.linspace(min_x, max_x, 100)
params = norm.pdf(overlay, mean, sd)
plt.plot(overlay, params, 'r')
```
