## Amortization 

### Term used to describe retiring a long term debt
### Frequently associated with mortgage loans
### Table that contains loan balance, interest and principal payments over the life of the loan
### We can build a small application in the form of a function that allows us to evaluate loans under various scenarios


```python
import numpy as np
import pandas as pd
import numpy_financial as finance
import matplotlib.pyplot as mat
%matplotlib inline
```

```python
pv = 360000
rate = 0.05875
term = 30
pmt = finance.pmt(rate/12, term * 12, -pv)
print("Value function: ", pmt)

## Generate an amortization table, with a lenght of 361 rows
amort_table = pd.DataFrame(
    {'Balance': np.zeros(term * 12 + 1),
     'Interest': np.zeros(term * 12 + 1 ),
     'Principal': np.zeros(term * 12 + 1),
     'Cum_Int':np.zeros(term * 12 + 1)})
amort_table.loc[0,'Balance'] = pv
amort_table.loc[0,'Interest'] = amort_table.loc[0,'Balance'] * rate / 12
amort_table.loc[0,'Principal']= pmt - amort_table.loc[0,'Interest']
amort_table.loc['Cum_Int'] = amort_table['Interest'].cumsum()
```
