## Internal Rate Of Return

### Somewhat flawed method of evaluation investments
### Only works in specific situations
### Returns a cost of capital that makes NPV approximately 0
### Realiable only when there is one net negative cash outflow

```python
import numpy as np
import numpy_financial as nf
```

```python
initial_investment = 100000
cost_of_capital = .12
years = np.arange(1,6)
cashflows = np.ones(5) * 30000

discounted_cashflows = cashflows / (1 + cost_of_capital) ** years
nvp = sum(discounted_cashflows) - initial_investment
print(nvp)

## One way to do the IRR
cost_of_capital = nf.irr([-initial_investment, 30000, 30000, 30000, 30000, 30000])
print(cost_of_capital)

discounted_cashflows = cashflows / (1 + cost_of_capital) ** years
npv = initial_investment - sum(discounted_cashflows)
print(npv)

```
