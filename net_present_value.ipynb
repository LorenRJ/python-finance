## Net Present Value

### Commonly used to evaluate capital investment projects
### In bussines settings there are often multiple ways to invest in the bussines
### Should you undertake a distribution channel upgrade or invest in information technology??
### Net presentg value evaluation allows the direct comparison of projects 
### Relies heavily on uncertain values such as future benefits and cost of capital

```python
import numpy as np
```

```python
initial_investment = 100000
cost_of_capital = .12
years = np.arange(1,6)
cashflows = np.ones(5) * 30000

## Every year with less money each year
discounted_cashflows = cashflows / (1 + cost_of_capital) ** years
print(discounted_cashflows)

## NPV
npv = np.sum(discounted_cashflows) - initial_investment
print("NPV : ", npv)
```
