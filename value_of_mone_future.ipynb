### Value of Money in the Future

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## basic example
r = 0.05
T = 1
pv = 50000
fv = pv * (1+r)**T

def future_value(r, T, pv):
    return '${:,.2f}'.format(round(pv * (1 + r)**T, 2))

future_value(r, 2, pv)


## Ending balances

future_values = np.ones(20) * 50000
for future_value in range(1, len(future_values)):
    future_values[future_value] = round(future_values[future_value] * (1 + r) ** future_value, 2)
print(future_values)

## Make a plot 
plt.plot(future_values)
plt.title("Future Value Over Time")
plt.grid("--")
plt.xticks([0,5,10,15,20])
```

```python

np.random.normal(.095, .185)
edge_values = future_values
for t in range(len(edge_values)):
    if(t <= 0 ):
        edge_values[t] = edge_values[t]
    else:
        random_rate = np.random.normal(.095, .185)
        edge_values[t] = round(edge_values[t-1] * (1 + random_rate),2)
plt.plot(edge_values)
plt.title("Values whith a random interest each year")
plt.grid(":")
plt.xticks([0,5,10,15,20])
```
